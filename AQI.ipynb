#### Null  Handling

```python
df.isnull().sum()
```

In number_of_monitoring_stations there are 141 null values so we can assume that
there must be 1 monitoring_stations should be there to find out calculation
so we replace null by 1.


```python
df["number_of_monitoring_stations"] = df["number_of_monitoring_stations"].fillna(1)
```

```python
df["number_of_monitoring_stations"]=df["number_of_monitoring_stations"].replace({0:1})
```

```python
df.isnull().sum()
```

```python
def set_season(month) :
    if month in [12 , 1, 2] :
        return "Winter"
    elif month in [3,4,5]:
        return "Summer"
    elif month in [6,7,8,9] :
        return "Monsoon" 
    else :
        return "Post Monsoon"

df["season"] = df["month"].apply(set_season)
```

```python
df
```

```python
def validate_aqi_category(aqi):
    if aqi <= 50 :
        return "Good"
    elif aqi <= 100 :
        return "Satisfactory"
    elif aqi <= 200 :
        return "Moderate"
    elif aqi <= 300 :
        return "Poor"
    elif aqi <= 400 :
        return "Very Poor"
    else :
        return "Severe"

df["aqi_parameter"] = df["aqi_value"].apply(validate_aqi_category)
    
```

```python
df["aqi_status_check"] = df["air_quality_status"] == df["aqi_parameter"]
```

```python
df["aqi_status_check"].value_counts()
```

```python
AS clearly shown From above figures that the original data set holds some wrong values in column
"air_quality_status" and we created one new column in which all value are right ,so we can delete
"air_quality_status".
```

```python
df = df.drop(columns = "air_quality_status")
```

```python
df = df.drop(columns = ["unit","aqi_status_check"])
```

```python
def determine_category(pollutant):
    if pollutant in ["PM2.5","PM10"]:
                 return 'Particulate'
    elif pollutant in ['CO', 'SO2', 'NO2','O3']:
                 return 'Gaseous'
    else:
                 return 'Secondary' 


df["pollutant Category"] = df["prominent_pollutants"].apply(determine_category)
```
